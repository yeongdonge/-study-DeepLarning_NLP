<p><b>계단함수</b></p>
<p>이제 거의 사용되지 않지만, 가장 처음 접하게 되는 활성화 함수</p>

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image
```

```python
def step(x):
    return np.array(x > 0, dtype=np.int)
x = np.arange(-5.0, 5.0, 0.1)
y = step(x)
plt.title('Step Function')
plt.plot(x,y)
plt.show()
```

<p><b>시그모이드 함수와 기울기 소실</b></p>
<p>시그모이드 함수를 이용한 아래의 신경망이 있다고 가정</p>

```python
Image("simple-neural-network.png")
```

<p><b>위 인공 신경망의 학습 과정</b></p>
<ul>
    <li>입력에 대해서 순전파를 연산</li>
    <li>순전파 연산을 통해 나온 예측값과 실제값의 오차를 손실함수를 통해 계산</li>
    <li>손실을 미분을 통해서 기울기를 구함</li>
    <li>이를 통해 역전파를 수행</li>
</ul>
<b>시그모이드 함수의 문제점은 미분을 해서 기울기를 구할 때 발생</b>

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)

plt.plot(x,y)
plt.plot([0,0],[1.0,0.0], ':')
plt.title('Sigmoid Function')
plt.show()
```

<p> 시그모이드 함수의 출력값이 0 또는 1에 가까워지면 기울기가 완만해지는 모습을 볼 수 있음</p>
<p> 완만해지는 구간을 주황색, 그렇지 않은 구간을 초록색으로 칠함</p>
<p> 주황색 부분의 기울기를 계산하면, 0에 가까운 아주 작은 값이 나오게 됨</p>
<p> 역전파 과정에서, 0에 가까운 아주 작은 기울기가 곱해지면, 앞단에는 기울기가 잘 전달되지 않음</p>
<p> 이러한 현상을 <b>기울기 소실</b>문제라고 함</p>
<p> 결론적으로 시그모이드 함수를 은닉층에서 사용하는 것은 지양됨</p>
