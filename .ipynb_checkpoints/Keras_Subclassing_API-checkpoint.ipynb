```python
# 케라스의 Sequential API로 구현했던 선형 회귀를 SUbclassing API로 구현
import tensorflow as tf

class LinearRegression(tf.keras.Model):
    def __init__(self):
        super(LinearRegression, self).__init__()
        self.linear_layer = tf.keras.layers.Dense(1, input_dim=1, activation='linear')
    
    def call(self, x):
        y_pred = self.linear_layer(x)
        
        return y_pred
```

```python
model = LinearRegression()
```

```python
X = [1,2,3,4,5,6,7,8,9]
y = [11,22,33,44,53,66,77,87,95]

# sgd는 경사 하강법을 의미, 학습률(learning rate, 1r)은 0.01
sgd = tf.keras.optimizers.SGD(lr=0.01)

# 손실 함수(Loss function)은 평균제곱오차 mse를 사용
model.compile(optimizer=sgd ,loss='mse',metrics=['mse'])

# 주어진 X와 y데이터에 대해서 오차를 최소화하는 작업을 300번 시도
model.fit(X,y, batch_size=1, epochs=300, shuffle=False)
```

<b> 클래스(class) 형태의 모델은 tf.keras.Model을 상속, 그리고 init()에서 모델의 구조와 동적을 정의하는 생성자를 정의<br>
    객체가 생성될 때 자동으로 호출되며, super() 함수를 부르면 여기서 만든 클래스는<br>
    tf.keras.Model 클래스의 속성들을 가지고 초기화가 됨<br>
    call() 함수는 모델이 데이터를 입력받아 에측값을 리턴하는 포워드(foward) 연산을 진행<br>
    H(x) 식에 입력 x로부터 예측된 y를 얻는 것을 forward 연산이라고 함
